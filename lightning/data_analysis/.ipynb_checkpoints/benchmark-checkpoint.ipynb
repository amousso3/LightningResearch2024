# Benchmark

We begin with a simple analysis of the lightning variables. The lightning data is the target, while the environmental variables will be the predictors. We will turn lightning into a binary variable and use Sci-Kit Learn to find the log odds for each variable over time. Code will be derived form Henri's work.

```python
import xarray as xr
import cdsapi
import netCDF4 as nc
from goes2go import GOES
import requests
from datetime import datetime, timedelta
import os
from concurrent.futures import ThreadPoolExecutor, as_completed
import numpy as np
```

```python
def romps(cape, precip):
    return cape * precip

def updraught_mass_flux(w, t, s_h,):
    p = 400*100 # Air Pressure
    R_v = 461.5 # Gas constant for water vapor
    R_d = 287.05 # Gas constant for Dry Air
    T_v = t * (1 + s_h * ((R_v/R_d) -1))
    rho = p / (R_d * T_v)
    F = w * rho # Updraught Mass Flux at 450 hPa
    return F

def ice_flux(q, w, t, s_h, c):
    """Returns the ICEFLUX parameterization of lightning from the 2014 Finney Study
        at 400 hPa.
    """
    F = updraught_mass_flux(w ,t,s_h) # Updraught Mass Flux at 400 hPa
    phi = (q * F) /c #ICEFlUX Variable
    mask = c < 0.01
    phi = phi.where(~mask, 0)
    return phi
```

```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Initialize an array to store coefficients (2 predictors, 24 hours)
coefficients = np.zeros((2, 24))

# Loop through each hour of the day
for hour in range(24):
    # Select data for the current hour
    lightning_hour = lightning_data.sel(time=lightning_data.time.dt.hour == hour)
    cape_precip_hour = cape_precip.sel(time=cape_precip.time.dt.hour == hour)
    iceflux_hour = iceflux.sel(time=iceflux.time.dt.hour == hour)

    # Flatten the dataarrays
    y_hour = lightning_hour.values.flatten()
    X_hour_cape_precip = cape_precip_hour.values.flatten()
    X_hour_iceflux = iceflux_hour.values.flatten()

    # Combine predictors into a single feature matrix
    X_hour = np.vstack([X_hour_cape_precip, X_hour_iceflux]).T

    # Remove NaN values
    valid_mask = ~np.isnan(X_hour).any(axis=1) & ~np.isnan(y_hour)
    X_hour = X_hour[valid_mask]
    y_hour = y_hour[valid_mask]

    # Ensure binary target (presence/absence of lightning)
    y_hour = (y_hour > 0).astype(int)

    # Skip if there's insufficient data for regression
    if len(y_hour) < 10:  # You can adjust this threshold
        coefficients[:, hour] = np.nan
        continue

    # Perform logistic regression
    model = LogisticRegression()
    model.fit(X_hour, y_hour)

    # Store coefficients for this hour
    coefficients[:, hour] = model.coef_[0]

# Coefficients are stored in (2, 24) array
```

```python
# Plot the coefficients as time series
hours = np.arange(24)
plt.figure(figsize=(10, 6))
plt.plot(hours, coefficients[0, :], label='Cape x Precip Coefficient')
plt.plot(hours, coefficients[1, :], label='ICEFLUX Coefficient')
plt.xlabel('Hour of Day')
plt.ylabel('Logistic Regression Coefficient')
plt.title('Hourly Logistic Regression Coefficients')
plt.legend()
plt.grid()
plt.show()
plt.savefig('log_regression_1.png')
```

Looking at the individual parameters has allowed me to make some observations and propose some ideas:
* ICEFLUX may have negative log odds (which means every unit increase leads to an $e^{o} - 1$ percent decrease where o is the log odds), not because it is a bad parameterization, but because it may overestimate lightning by a large amount. For example, cloud cover and precipitation have relatively negative log odds. This is seemingly wrong, however, physically speaking, cloud cover and rain occur at a significantly higher rate than lightning does. Although they are good estimates (since lightning usually does not occur without one or the other), it could simply be that these parameters would predict lightning where there isn't any most of the time. It is also important to remember that this is only within a day.
* CAPE and specific humidity have the most consistently high log odds. This makes physical sense.
